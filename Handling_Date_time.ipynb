# Working on dates with orders column

**first we have to convert date datatype to (date_time)**

```python
date.info()
```

```python
date["date"] = pd.to_datetime(date["date"])
```

```python
date.info()
```

# Extract Year

```python
date["Year"] = date["date"].dt.year
```

```python
date
```

# Extract Month and Month Name

```python
date["Month"] = date["date"].dt.month
```

```python
date
```

```python
date["Month_name"] = date["date"].dt.month_name()
```

```python
date
```

# Extract day

```python
date["Day"] = date["date"].dt.day
```

```python
date
```

# Extract day name

```python
date["Day_name"] = date["date"].dt.day_name()
```

```python
date
```

# Extract if day is weekend or not

```python
date["Weekend"] = np.where(date["Day_name"].isin(["Sunday","Saturday"]),"yes","no")
```
